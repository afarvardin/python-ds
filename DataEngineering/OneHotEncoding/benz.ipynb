```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('mercedes-benz.csv')
```

```python
top10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top10
```

```python
for label in top10:
    data[label] = np.where(data["X2"] == label, 1, 0)

# df['X2']
data[['X2'] + top10].head(5)
```

```python
from statistics import variance


def one_hot_top_x(df, cat_variables, top_x_labels):
    for label in top_x_labels:
        df[cat_variables+'_'+label] = np.where(data[cat_variables] == label, 1, 0)

data  = pd.read_csv('mercedes-benz.csv', usecols=['X1','X2','X3','X4','X5','X6'])
top10_X2 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]

one_hot_top_x(data, 'X2', top10_X2)
data.head(10)
```
